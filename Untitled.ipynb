```python
"""
Helmet Detection for Bike Riders â€“ Final Project
Author: Vadlapudi Varun Kumar
Model: YOLOv8 (Ultralytics)
Description:
    Detects whether a bike rider is wearing a helmet or not using
    a pretrained YOLOv8 model. Supports both image and folder input.
"""

from ultralytics import YOLO
import cv2
import os
import matplotlib.pyplot as plt
from pathlib import Path

# -------------------------------
# 1. Load YOLOv8 Model
# -------------------------------
print("Loading YOLO model...")
model = YOLO("yolov8n.pt")
print("Model loaded successfully!\n")

# -------------------------------
# 2. Input / Output Folders
# -------------------------------
INPUT_FOLDER = Path("input_images")     # Add your test images here
OUTPUT_FOLDER = Path("results")        # Output folder
OUTPUT_FOLDER.mkdir(exist_ok=True)

# Create input_images folder if missing
if not INPUT_FOLDER.exists():
    INPUT_FOLDER.mkdir()
    print("âš ï¸ 'input_images' folder was missing â€” created automatically!")
    print("âž¡ï¸ Add images inside 'input_images/' and run again.\n")
    exit()

# Allowed image types
IMAGE_EXTENSIONS = [".jpg", ".jpeg", ".png"]


# -------------------------------
# 3. Detection Function
# -------------------------------
def detect_helmet(image_path):
    """Runs YOLO detection and saves the output."""
    print(f"ðŸ”Ž Detecting helmet in: {image_path.name}")

    results = model.predict(source=str(image_path), save=False)

    for r in results:
        output_img = r.plot()
        save_path = OUTPUT_FOLDER / f"result_{image_path.name}"
        cv2.imwrite(str(save_path), output_img)
        print(f"âœ” Output saved: {save_path}")


# -------------------------------
# 4. Run detection on all images
# -------------------------------
def run_on_folder():
    image_files = [
        f for f in INPUT_FOLDER.iterdir()
        if f.suffix.lower() in IMAGE_EXTENSIONS
    ]

    if len(image_files) == 0:
        print("âš ï¸ No valid images found in 'input_images/' folder.")
        return

    for img_path in image_files:
        detect_helmet(img_path)

    print("\nðŸŽ‰ Detection Completed!")
    print(f"ðŸ“‚ All results saved inside: {OUTPUT_FOLDER}\n")


# -------------------------------
# 5. Display Sample Output
# -------------------------------
def show_output(sample_file):
    """Display a saved result image."""
    img = cv2.imread(str(sample_file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 6))
    plt.imshow(img)
    plt.title("Helmet Detection Output")
    plt.axis("off")
    plt.show()


# -------------------------------
# 6. Main Execution
# -------------------------------
if __name__ == "__main__":
    run_on_folder()

    output_images = list(OUTPUT_FOLDER.glob("result_*"))
    if output_images:
        print("ðŸ“¸ Showing sample output...")
        show_output(output_images[0])
    else:
        print("âš ï¸ No output images found to display.")
```
